```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_excel('/content/ecommerce_data.xlsx')
df.head(5)
```

```python
df.info()
```

```python
df['Avg. Session Length']=df['Avg. Session Length'].interpolate(direction='forward')
```

```python
df['Time on App']=df['Time on App'].interpolate(direction='forward')
```

```python
df['Time on Website']= df['Time on Website'].interpolate(direction='forward')
```

```python
df['Length of Membership']= df['Length of Membership'].interpolate(direction='forward')
```

```python
df['Yearly Amount Spent']= df['Yearly Amount Spent'].interpolate(direction='forward')
df['Yearly Amount Spent'].info()
```

```python
df['Avatar']=df['Avatar'].ffill()
```

```python
df.head(20)
```

```python
df.info()
```

```python
df.dropna()
```

```python
df_new=df[['Avg. Session Length', 'Time on App','Time on Website','Length of Membership','Yearly Amount Spent']]
df_new.dropna()
```

```python
sns.boxplot(df_new,orient='h')
```

```python
Q1 = df_new.quantile(0.25)
Q3 = df_new.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
df_new = df_new[~((df_new < (Q1 - (1.5 * IQR))) |(df_new > (Q3 + (1.5 * IQR)))).any(axis=1)]
```

```python
sns.boxplot(df_new,orient='h',width=.5)
```

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
num_std = std_scaler.fit_transform(df_new)
num_std_scaled_df = pd.DataFrame(num_std, columns=df_new.columns)
df_nu=num_std_scaled_df.dropna()
df_nu
```

```python
sns.pairplot(df_nu)
```

```python
m=df_nu.corr()
m
```

```python
sns.heatmap(m)
```

```python
df_nu.drop('Time on Website',axis=1)
```

```python
X=df_nu[['Avg. Session Length', 'Time on App','Length of Membership']]
Y=df_nu['Yearly Amount Spent']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
trainX,testX,trainy,testy=train_test_split(X,Y,test_size=0.2)
```

```python
trainy.shape
testy.shape
```

```python
reg=LinearRegression()
reg.fit(trainX,trainy)
```

```python
reg.coef_
```

```python
reg.score(trainX,trainy)
```

```python
reg.score(testX,testy)
```

```python

y_pred=reg.predict(testX)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(testy, y_pred)
rmse = np.sqrt(mse)
print("RMSE value: {:.4f}".format(rmse))
```

```python
from sklearn.metrics import r2_score
print ("R2 Score value: {:.4f}".format(r2_score(testy, y_pred)))
```

```python
import matplotlib.pyplot as plt

plt.scatter((trainy), reg.predict(trainX) - trainy, color = 'red', label = 'Train data')
plt.scatter((testy), reg.predict(testX) - testy, color = 'blue', label = 'Test data')
plt.hlines(xmin = 0, xmax = 50, y = 0, linewidth = 3)
plt.title('Residual errors')
plt.legend(loc = 4)
plt.show()
```

```python

from sklearn.model_selection import cross_validate,cross_val_score,KFold
folds = KFold(n_splits = 5, shuffle = True, random_state = 100)
result=cross_val_score(estimator=reg,X=trainX,y=trainy,cv=folds,scoring='r2')
result
```

```python
from sklearn.metrics import get_scorer_names
print(get_scorer_names())
```

```python
# specify model
from sklearn.feature_selection import RFE
from sklearn.model_selection import GridSearchCV
lm = LinearRegression()
lm.fit(trainX, trainy)
rfe = RFE(lm)

# set up GridSearchCV()
params= [{'n_features_to_select': list(range(1, 5))}]
model_cv = GridSearchCV(estimator = rfe,
                        param_grid = params,
                        scoring= 'r2',
                        cv = folds,
                        verbose = 1,
                        return_train_score=True)

# fit the model
model_cv.fit(trainX, trainy)
```

```python
model_cv.best_params_
```

```python
model_cv.feature_names_in_
```

```python
model_cv.best_score_
```

```python
model_cv.cv_results_
```

```python
cv_results = pd.DataFrame(model_cv.cv_results_)
cv_results
```

```python
plt.figure(figsize=(16,6))

plt.plot(cv_results["param_n_features_to_select"], cv_results["mean_test_score"])
plt.plot(cv_results["param_n_features_to_select"], cv_results["mean_train_score"])
plt.xlabel('number of features')
plt.ylabel('r-squared')
plt.title("Optimal Number of Features")
plt.legend(['test score', 'train score'], loc='upper left')
```

```python
# final model
n_features_optimal = 3

lm = LinearRegression()
lm.fit(trainX, trainy)

rfe = RFE(lm, n_features_to_select=n_features_optimal)
rfe = rfe.fit(trainX, trainy)

# predict prices of X_test
y_pred = lm.predict(testX)
r2 = r2_score(testy, y_pred)
print(r2)
```
